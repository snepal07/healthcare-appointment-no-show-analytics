# Executive Business Insights

## Key Findings

### 1. Overall Attendance
Approximately 80% of patients attended their appointments, while around 20% missed them.

### 2. Waiting Time
Patients with longer waiting periods showed substantially higher no-show rates. Long scheduling delays appear to increase appointment abandonment.

### 3. SMS Reminders
Patients who received SMS reminders generally attended appointments more frequently than those who did not, suggesting reminder systems improve attendance.

### 4. Scholarship
Patients receiving scholarships exhibited a slightly higher no-show rate, indicating potential socioeconomic barriers.

### 5. Age
Younger patients were generally more likely to miss appointments than older patients.

### 6. Medical Conditions
Hypertension and diabetes showed only modest relationships with attendance. Alcoholism had minimal impact.

### 7. Neighbourhood
Several neighbourhoods consistently experienced higher no-show rates, highlighting opportunities for targeted outreach programs.

## Business Recommendations

- Increase SMS reminder campaigns.
- Reduce patient waiting times.
- Focus reminder efforts on high-risk neighbourhoods.
- Develop interventions for younger patients.
- Prioritize operational improvements for appointments scheduled more than one month in advance.
